# Perform a brief visualization of the TBH Sparse Matrices, with provided data sets.
## Import Numpy and Matplotlib
## And define a helper function which does a sub-sum of a provided matrix.

```python
import numpy as np
import matplotlib.pyplot as plt

def sum_11_submatrix(mat,n,m):
    partial=0
    for i in range(11):
        for j in range(11):
            partial+=mat[11*n+i][11*m+j]
    
    return partial
```

## This TBH was generated using a 2x2 unit vector base (4 11-orbital 'atoms' on each monolayer, 8 atoms total in the bilayer system).

```python
TBH = np.load('TBH.2x2')

fig = plt.figure(1,figsize=(3,3),dpi=150)
ax = fig.add_subplot(111)
ax.matshow(np.abs(TBH),cmap='Greys')
plt.show()

NN = 8
renormalized = np.zeros((NN,NN),dtype='float')

#Check to see that the loading occurred correctly
print(TBH.shape)

#Generate the `renormalized' matrix
for i in range(NN):
    for j in range(NN):
        renormalized[i][j] = sum_11_submatrix(np.abs(TBH),i,j)

# Make the other figure
fig=plt.figure(1,figsize=(3,3),dpi=150)
ax=fig.add_subplot(111)
ax.matshow(renormalized,cmap='Greys')

ax.set_xticklabels(range(0,9))
ax.set_yticklabels(range(0,9))

plt.show()


```
